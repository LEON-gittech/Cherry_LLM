```python
import matplotlib.pyplot as plt

# 
retrieval_amounts = [1000, 2000, 3000, 4000, 5000]
# code = [0.3256, 0.3292, 0.3353, 0.3536, 0.3658]
code = [0.3353, 0.3536, 0.3110, 0.3293, 0.3658]
med = [0.712, 0.718, 0.721, 0.728, 0.734]
fin = [0.6716, 0.6922, 0.7004, 0.7062, 0.7124]
math = [0.5002, 0.5083, 0.5155, 0.5231, 0.5246]

code = [x * 100 for x in code]
med = [x * 100 for x in med]
fin = [x * 100 for x in fin]
math = [x * 100 for x in math]

# Domain Coverage 
code_coverage = [0.9183, 0.9343, 0.9547, 0.9666, 0.9767]
med_coverage = [0.7876, 0.8287, 0.8593, 0.8828, 0.9348]
fin_coverage = [0.9344, 0.9561, 0.9667, 0.9736, 0.9815]
math_coverage = [0.8625, 0.8877, 0.9036, 0.9152, 0.9320]

# 
fig, axs = plt.subplots(2, 4, figsize=(20, 8)) 

# 
axs[0, 0].plot(retrieval_amounts, code, marker='o', linestyle='-', color='b', label='HumanEval(Pass@1)')
axs[0, 0].set_ylabel('Performance(%)')
axs[0, 0].legend()
axs[0, 0].grid(True)

axs[0, 1].plot(retrieval_amounts, med, marker='^', linestyle='-', color='r', label='MMLU-Med(Acc)')
axs[0, 1].set_ylabel('Performance(%)')
axs[0, 1].legend()
axs[0, 1].grid(True)

axs[0, 2].plot(retrieval_amounts, fin, marker='s', linestyle='-', color='g', label='FPB(Acc)')
axs[0, 2].set_ylabel('Performance(%)')
axs[0, 2].legend()
axs[0, 2].grid(True)

axs[0, 3].plot(retrieval_amounts, math, marker='x', linestyle='-', color='y', label='GSM8K(Exact Match)')
axs[0, 3].set_ylabel('Performance(%)')
axs[0, 3].legend()
axs[0, 3].grid(True)

# 
axs[1, 0].plot(retrieval_amounts, code_coverage, marker='o', linestyle='-', color='b', label='Code')
axs[1, 1].set_xlabel('Retrieval Amount')
axs[1, 0].set_ylabel('Domain Coverage')
axs[1, 0].legend()
axs[1, 0].grid(True)

axs[1, 1].plot(retrieval_amounts, med_coverage, marker='^', linestyle='-', color='r', label='Medical')
axs[1, 1].set_xlabel('Retrieval Amount')
axs[1, 1].set_ylabel('Domain Coverage')
axs[1, 1].legend()
axs[1, 1].grid(True)

axs[1, 2].plot(retrieval_amounts, fin_coverage, marker='s', linestyle='-', color='g', label='Financial')
axs[1, 2].set_xlabel('Retrieval Amount')
axs[1, 2].set_ylabel('Domain Coverage')
axs[1, 2].legend()
axs[1, 2].grid(True)

axs[1, 3].plot(retrieval_amounts, math_coverage, marker='x', linestyle='-', color='y', label='Mathematical')
axs[1, 3].set_xlabel('Retrieval Amount')
axs[1, 3].set_ylabel('Domain Coverage')
axs[1, 3].legend()
axs[1, 3].grid(True)

# 
plt.tight_layout()
plt.savefig('res/retrieval_num.eps', format='eps')
# 
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
k_values = [10, 20, 40, 80]
code = [0.9766, 0.9720, 0.9625, 0.9782]
med = [0.9348, 0.9478, 0.9466, 0.9618]
fin = [0.9815, 0.9814, 0.9819, 0.9813]
math = [0.9320, 0.9348, 0.9344, 0.9337]

# 1x4 grid of subplots
fig, axs = plt.subplots(1, 4, figsize=(20, 5))  # 1 row, 4 columns

#  - Code Domain
axs[0].plot(k_values, code, marker='o', linestyle='-', color='b')
axs[0].set_xlabel('k')
axs[0].set_ylabel('Domain coverage')
axs[0].set_title('Code')
axs[0].grid(True)

#  - Med Domain
axs[1].plot(k_values, med, marker='^', linestyle='-', color='r')
axs[1].set_xlabel('k')
axs[1].set_ylabel('Domain coverage')
axs[1].set_title('Medical')
axs[1].grid(True)

#  - Fin Domain
axs[2].plot(k_values, fin, marker='s', linestyle='-', color='g')
axs[2].set_xlabel('k')
axs[2].set_ylabel('Domain coverage')
axs[2].set_title('Financial')
axs[2].grid(True)

#  - Math Domain
axs[3].plot(k_values, math, marker='x', linestyle='-', color='y')
axs[3].set_xlabel('k')
axs[3].set_ylabel('Domain coverage')
axs[3].set_title('Mathematical')
axs[3].grid(True)

# 
plt.tight_layout()
plt.savefig('res/k.eps', format='eps')
# 
plt.show()
```
